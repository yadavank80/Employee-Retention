# 3. Distributions of categorical features

```python
df.describe(include='object')
```

```python
plt.figure(figsize=(10,10))
categorical = df.columns[df.dtypes == 'object']

for col in categorical:
    sns.countplot(data = df, y=col)
    plt.show()
```

## Conclusion derived from analysis 3: 
 * We have missing observations in feature <code style="color:steelblue">department</code>
 * We have **12 departments** out of which around **28% of total employees** belong to <code style="color:steelblue">sales</code> department followed by <code style="color:steelblue">engineering</code> and <code style="color:steelblue">support</code>.
 * **48.78** employees belong to **low salary** class.
 * We have more data about about currently employeed employees(76.20%) than employees who have left.

# 4. Segmentations

```python
plt.figure(figsize=(10,10))
sns.violinplot(data = df, y = 'status', x = 'satisfaction')
plt.show()
```

```python
# Segment last_evaluation by status
plt.figure(figsize=(10,10))
sns.violinplot(data = df, y = 'status', x = 'last_evaluation')
plt.show()
```

```python
# Segment by status and display the means within each class
df.groupby('status').mean()
```

```python
# Scatterplot of satisfaction vs. last_evaluation
plt.figure(figsize=(10,10))
sns.lmplot(data=df, x='satisfaction', y='last_evaluation', hue='status')
plt.show()
```

```python
# Scatterplot of satisfaction vs. last_evaluation, only those who have left
plt.figure(figsize=(10,10))
sns.lmplot(data=df[df.status == 'Left'], x='satisfaction', y='last_evaluation')
plt.show()
```

## Conclusion derived from analysis 4: 
 * Employees who left had low average satisfaction score (0.44) (refer cell 38 output)
 * We can classify employees who left into three different categories:<br>
     i) Unsatisfied employees with <code style="color:steelblue">satisfaction</code> score < 0.2 <br>
     ii) Underperforming employees with <code style="color:steelblue">last_evaluation</code> < 0.5<br>
     iii) Top performing employees with <code style="color:steelblue">satisfaction</code> score > 0.7 and <code style="color:steelblue">last_evaluation</code> > 0.8
